```python
import requests
import os
import json

url = "http://api.jingtw.tk/liver_voice/api/addition"

def download_file(url, filename):
    with open(filename, "wb") as file:   # open in binary mode
        response = requests.get(url)  
        #print(response.content)# get request
        file.write(response.content)      # write to file
        
def get_file_url(content,noise_scale,noise_scale_w,speed):

  payload = json.dumps({
    "data": [
      content,
      "Liver",
      "中文",
      noise_scale,
      noise_scale_w,
      speed,
      None
    ]
  })
  headers = {
    'Content-Type': 'application/json'
  }

  response = requests.request("POST", url, headers=headers, data=payload)

  res = json.loads(response.text)
  voice_path = res['data'][1]['name'].replace('\\','//')
  file_url = f"http://api.jingtw.tk/liver_voice/file={voice_path}"
  return file_url
#download_file(get_file_url("早上好中國，現在我有冰淇淋"), "test.wav")
```

```python
noise_scale = 0.8
noise_scale_w= 0.6
speed = 10
```

```python
#for each file content in Story folder

for file in os.listdir('./Story/'):
    if file.startswith('.'):
        continue
    title = file.split('_')[0]
    print(title)
    
    with open('./Story/'+file, 'r') as f:
        content = f.readlines()
        #delete \n
        content = [x.strip() for x in content]
        #delete empty line
        content = [x for x in content if x]
        clean_content = []  
        for i in content:
            clean_content.append([x for x in i.split('。') if x])
    
    for i in clean_content:
        for j in i:
            
            if not (os.path.exists(f"./Voice/{title}")):
                os.makedirs(f"./Voice/{title}")
            clean_content_index = "{:2}".format(clean_content.index(i)).replace(' ','0')
            print(f"{clean_content_index}_{i.index(j)}.wav")
            if(os.path.exists(f"./Voice/{title}/{clean_content.index(i)}_{i.index(j)}.wav")):
                print("File exist")
                continue
            download_file(get_file_url(j,noise_scale,noise_scale_w,speed), f"./Voice/{title}/{clean_content_index}_{i.index(j)}.wav")
```

```python
import pyaudio  
import wave
import time

chunk = 1024

#play audio

def playStory(storyname):
    for file in sorted(os.listdir(f'{storyname}/')):
  
        #open a wav format music  
        f = wave.open(rf"{storyname}/{file}","rb")  
        #instantiate PyAudio  
        p = pyaudio.PyAudio()  
        #open stream  
        stream = p.open(format = p.get_format_from_width(f.getsampwidth()),  
                        channels = f.getnchannels(),  
                        rate = f.getframerate(),  
                        output = True)  
        #read data  
        data = f.readframes(chunk)  
        
        #play stream  
        while data:  
            stream.write(data)  
            data = f.readframes(chunk)  
        
        time.sleep(0)
        #stop stream  
    stream.stop_stream()  
    stream.close()  
        
        #close PyAudio  
    p.terminate() 
    
playStory("./Voice/勇敢的騎士艾倫/")
```
